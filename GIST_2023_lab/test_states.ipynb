```python
from qiskit import QuantumCircuit
```

```python
from qiskit.circuit.library import QFT
from qiskit.quantum_info import Statevector
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
qc = QuantumCircuit(4, name="test")

qc.append(QFT(4), [0, 1, 2, 3])
qc.x(1)
qc.rz(36*np.pi/32, 1)
qc.append(QFT(4).inverse(), [0,1,2,3])

#qc.draw("mpl")
std_vec = Statevector(qc)
c_vec = np.array(std_vec)
p_state = (c_vec * c_vec.conjugate()).astype(float)
plt.plot(np.arange(16), p_state)
```
