```python
import pennylane as qml
from math import pi
```

```python
wires = 3
dev = qml.device("default.qubit", wires=wires)
```

```python
@qml.qnode(dev)
def circuit():
    qml.CNOT(wires=[2, 0])
    qml.CNOT(wires=[1, 0])
    qml.RZ(pi/3, 0)
    qml.CNOT(wires=[1, 0])
    qml.CNOT(wires=[2, 0])
    return qml.probs(wires=[0, 1, 2])
fig, ax =  qml.draw_mpl(circuit)()
fig.show()
```

```python
circuit()
```
