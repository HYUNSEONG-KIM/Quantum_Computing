```python
import numpy as np 
from numpy import kron
from itertools import combinations
from functools import reduce

import matplotlib.pyplot as plt
```

Complete graph: All vertex pair is connected, $(v_i,v_j) \in E, v_i, v_j \in V$ of $G= (V, E)$.

Max-cut of complete graph.
- Total number of edge of graph, $G$: $\begin{pmatrix} ||V|| \\ 2\end{pmatrix} = \frac{1}{2} ||V|| (||V||-1)$
- Max-cut: $||V||^2/4$
- Degeneracy state: $\begin{pmatrix} ||V|| \\ ||V||/2\end{pmatrix}$


- 4-vertex-complete-graph: $4^2/4 = 4$
- 8-vertex-complete-graph: $8^2/4 = 16$

$$H = -\frac{1}{2}\sum_{i,j} (I-S_i S_j)$$

```python
qubits= 8
I = np.eye(2)
Z = np.diag([1, -1])
```

```python
edges = list(combinations(range(8), 2))
# Construct Hmailtonian
H = len(edges)*np.eye(int(2**qubits)) 
for edge in edges:
    i, j = edge
    IZs = qubits*[I]
    IZs[i] = Z
    IZs[j] = Z
    H = H - reduce(kron, IZs)
H *= 0.5
```

```python
plt.imshow(H)
```

```python
plt.plot(np.diag(H), "o", markersize=1)
```
