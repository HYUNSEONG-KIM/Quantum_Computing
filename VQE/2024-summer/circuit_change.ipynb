```python
import qiskit
from qiskit import QuantumCircuit
from qiskit
import matplotlib.pyplot as plt

%matplotlib inline

# Create a quantum circuit
qc = QuantumCircuit(3)
qc.h(0)
qc.cx(0, 1)
qc.cx(1, 2)
qc.cx(0, 2)
qc.measure_all()

# Draw the circuit and get the Matplotlib figure
# Define the custom style for the drawer
custom_style = {
    'gate_color': {
        'cx': 'red',  # Change the color of the CNOT gate to red
    },
    "backgroundcolor": "lightgreen"
}

# Draw the circuit with the custom style
fig = qc.draw(output='mpl', style=custom_style, interactive=False)

ax = fig.gca()  # Get the current axes

# Iterate over the text elements in the plot and change the color of the third cx gate
gate_count = 0

for text in ax.texts:
    print(text)
    if text.get_text() == '⊕':  # This is how 'cx' gate is represented in Qiskit visualization
        gate_count += 1
        if gate_count == 3:  # Change the color of the third cx gate
            for sibling in text.get_window_extent().inverse_transformed(ax.transData).get_points():
                # Set the color of the CNOT gate symbol (⊕)
                rect = plt.Rectangle(sibling, 0, 0, linewidth=1, edgecolor='red', facecolor='red')
                ax.add_patch(rect)
            text.set_color('red')

# Show the modified circuit
plt.show()

```
