```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

G.add_edge("YY", "ZX")
G.add_edge("YY", "XZ")
G.add_edge("YY", "IY")
G.add_edge("YY", "YI")

G.add_edge("ZX", "XZ")
G.add_edge("YI", "IY")
```

```python
G2 = nx.Graph()

G2.add_edge("YY", "ZX", weight= 3/2)
G2.add_edge("YY", "XZ", weight= 3/2)
G2.add_edge("YY", "IY", weight= 1)
G2.add_edge("YY", "YI", weight= 1)

G2.add_edge("ZX", "XZ", weight=1/2)
G2.add_edge("ZX", "IY", weight=1/2)
G2.add_edge("ZX", "YI", weight=3/2)

G2.add_edge("YI", "IY", weight=2)
G2.add_edge("YI", "XZ", weight=1/2)
G2.add_edge("IY", "XZ", weight=3/2)
```

```python
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(10, 5))
pos1= nx.circular_layout(G, scale=0.2)

nodes1 = nx.draw_networkx_nodes(
    G, pos=pos1, 
    node_size = 700,
    node_color="white",
    label = G.nodes,
    linewidths = 2,
    edgecolors = "black",
    ax=ax1
    )
labels1 = nx.draw_networkx_labels(G, pos1, ax=ax1)
#edge_labels_mark = nx.get_edge_attributes(G,'weight')
edges1 = nx.draw_networkx_edges(G, pos1, ax=ax1)
#edge_labels = nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels_mark )

pos2 = nx.circular_layout(G2, scale=0.2)
nodes2 = nx.draw_networkx_nodes(
    G2, pos=pos2, 
    node_size = 700,
    node_color="white",
    label = G2.nodes,
    linewidths = 2,
    edgecolors = "black",
    ax=ax2
    )
labels2 = nx.draw_networkx_labels(G2, pos2, ax=ax2)
edge_labels_mark2 = nx.get_edge_attributes(G2,'weight')
edges2 = nx.draw_networkx_edges(G2, pos2, ax=ax2)
edge_labels2 = nx.draw_networkx_edge_labels(G2, pos2, 
                                            edge_labels=edge_labels_mark2, ax=ax2)
plt.savefig("degenerate.pdf", format="pdf", bbox_inches="tight")
plt.draw()
```
